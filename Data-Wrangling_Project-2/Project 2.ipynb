# Step 9: Handle Missing Values & Outliers (Final Data)

```python
# Shape of final_data
print("Shape of final_data:", final_data.shape)
```

```python
# Check for missing values
missing_values = final_data.isnull().sum()
print("\nMissing Values:\n", missing_values[missing_values > 0])
```

```python
# Identify all columns with numeric types (including floats and ints)
numeric_cols = final_data.select_dtypes(include=[np.number]).columns.tolist()

# Identify all object columns
object_cols = final_data.select_dtypes(include=['object']).columns.tolist()

print("Numeric Columns:", numeric_cols)
print("Object Columns:", object_cols)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# ==========================
# Handling Missing Values
# ==========================
# I will fill missing values in numeric columns with the mean value
if numeric_cols:  
    imputer_num = SimpleImputer(strategy='mean')  
    final_data[numeric_cols] = imputer_num.fit_transform(final_data[numeric_cols])  # Fill missing values
    print("Missing values in numeric columns have been filled with the mean.")

# I will fill missing values in object columns with the most frequent value (mode)
if object_cols:  
    imputer_cat = SimpleImputer(strategy='most_frequent')  
    final_data[object_cols] = imputer_cat.fit_transform(final_data[object_cols])  # Fill missing values
    print("Missing values in object columns have been filled with the most frequent value.")
```

```python
# ==========================
# Identify Outliers
# ==========================
for col in numeric_cols:
    # Calculate Q1, Q3, and IQR
    Q1 = final_data[col].quantile(0.25)
    Q3 = final_data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find outliers
    outliers = final_data[(final_data[col] < lower_bound) | (final_data[col] > upper_bound)]

    # Print outliers
    print(f"Outliers in {col}:\n", outliers)
```

# Step 10: Check Skewness and Correlation of the final_data

```python
# ==========================
# 3. Check Skewness
# ==========================
skewness = final_data[numeric_cols].skew()
skewness_df = pd.DataFrame(skewness, columns=['Skewness']).reset_index()
skewness_df.columns = ['Column', 'Skewness']  # Rename columns for clarity

# Display the skewness DataFrame
print("Skewness DataFrame:\n", skewness_df)

import matplotlib.pyplot as plt
import seaborn as sns

# ==========================
# Visualize Distribution with Box Plot
# ==========================
plt.figure(figsize=(12, 6))
sns.boxplot(data=final_data[numeric_cols])
plt.title('Box Plot of Numeric Columns')
plt.xticks(rotation=45)
plt.show()
```

```python
# ==========================
# Check Correlation
# ==========================
correlation_matrix = final_data[numeric_cols].corr()

# Display the correlation matrix as a DataFrame
print("Correlation Matrix:")
correlation_matrix
```

```python
# Visualize correlation
plt.figure(figsize=(12, 8))  # Increase figure size for better readability
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', 
            square=True, cbar_kws={"shrink": .8}, linewidths=0.5)

plt.title('Correlation Heatmap', fontsize=16)
plt.xticks(fontsize=12, rotation=45)  # Adjust x-axis labels
plt.yticks(fontsize=12, rotation=0)   # Adjust y-axis labels
plt.tight_layout()
plt.show()
```
